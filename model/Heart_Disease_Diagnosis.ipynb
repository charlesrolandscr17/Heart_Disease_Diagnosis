```python
data = pd.read_csv('Heart Disease dataset.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('target',axis=1)
y = data['target']
```

```python
x.head()
```

```python
y[:5]
```

```python
import tensorflow as tf
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
checkpoint_filepath = 'ckpt/checkpoint.model.keras'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_filepath,
    monitor='val_loss',
    mode='min',
    save_best_only=True)
```

```python
model_0 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1)
])

model_0.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['mae'])

history = model_0.fit(x_train,y_train,epochs=100, callbacks=[model_checkpoint_callback], validation_data=(x_test,y_test))
```

```python
model_0.load_weights(checkpoint_filepath)
```

```python
model_0.evaluate(x_test,y_test)
```
